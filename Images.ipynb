```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
random_noise = np.random.rand(640, 640)
random_noise
```

```python
fig, ax = plt.subplots()
ax.imshow(np.asarray(random_noise), cmap="Greys")
```

```python
rgb_noise = np.random.rand(640, 640, 3)
fig, axes = plt.subplots(ncols=3)
axes[0].imshow(rgb_noise[:, :, 0], cmap="Reds")
axes[1].imshow(rgb_noise[:, :, 1], cmap="Greens")
axes[2].imshow(rgb_noise[:, :, 2], cmap="Blues")
```
